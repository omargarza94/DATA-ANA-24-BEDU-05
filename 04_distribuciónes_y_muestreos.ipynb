```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Ejemplo_01_Costos_Proyectos.csv')
df.head()
```

```python
costos = df['Costo_Total']
```

```python
tamano_muestra = 30
numero_muestras = 1000

medias_muestralas = []

for _ in range(numero_muestras):
  muestra = np.random.choice(costos, size=tamano_muestra, replace=True)
  medias_muestralas.append(np.mean(muestra))

medias_muestralas = np.array(medias_muestralas)


plt.figure(figsize=(8,6))
sns.histplot(medias_muestralas, kde=True, color='skyblue')
plt.title('Distribucion muestra de la media de costos')
plt.xlabel('Media de costos')
plt.ylabel('Frecuencia')

plt.show()
```
